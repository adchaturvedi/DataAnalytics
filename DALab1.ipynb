**2.Cleaning the data**

Counting missing values in dataset

```python
data.isnull().sum()
```

Deleting the unwanted columns

```python
data = data.drop(['INCIDENT_DATE'],axis=1)
```

```python
data = data.drop(columns=['POPULATION_GROUP_CODE','POPULATION_GROUP_DESC','MULTIPLE_OFFENSE','MULTIPLE_BIAS'])
```

Replacing empty cells in OFFENDER_ETHNICITY column with 'Unknown'

```python
data["OFFENDER_ETHNICITY"].fillna('Unknown', inplace = True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

Replacing null values in columns ADULT_VICTIM_COUNT,JUVENILE_VICTIM_COUNT,ADULT_OFFENDER_COUNT,JUVENILE_OFFENDER_COUNT,TOTAL_INDIVIDUAL_VICTIMS with mean values of respective columns

```python
x = data["ADULT_VICTIM_COUNT"].mean()
data["ADULT_VICTIM_COUNT"].fillna(x, inplace = True)
```

```python
x = data["JUVENILE_VICTIM_COUNT"].mean()
data["JUVENILE_VICTIM_COUNT"].fillna(x, inplace = True)
```

```python
x = data["ADULT_OFFENDER_COUNT"].mean()
data["ADULT_OFFENDER_COUNT"].fillna(x, inplace = True)
```

```python
x = data["JUVENILE_OFFENDER_COUNT"].mean()
data["JUVENILE_OFFENDER_COUNT"].fillna(x, inplace = True)
```

```python
x = data["TOTAL_INDIVIDUAL_VICTIMS"].mean()
data["TOTAL_INDIVIDUAL_VICTIMS"].fillna(x, inplace = True)
```

```python
data.isnull().sum()
```

**Interpretations**

The top 10 states with highest crime rate

```python

state = data.groupby('STATE_ABBR', as_index= False).mean().sort_values('VICTIM_COUNT', ascending = False) 
state

state = state.head(10)
plt.figure(figsize = (16,6))
ax = sns.barplot(data= state, x= 'STATE_ABBR', y= 'VICTIM_COUNT',palette= 'magma')
plt.show()
```

*from the above representation we can observe that maximum hate crime has been recorded for the state of Nebraska*

The top 5 hate crimes in U.S.A.

```python

state = data.groupby('BIAS_DESC', as_index= False).mean().sort_values('VICTIM_COUNT', ascending = False) 
state

state = state.head(5)
plt.figure(figsize = (30,8))
ax = sns.barplot(data= state, x= 'BIAS_DESC', y= 'VICTIM_COUNT',palette= 'magma')
plt.show()
```

*by observing the mean victim count for each bias description we can conclude that the highest type of bias that lead to hate crime in U.S.A. was Anti-Lesbian(Female)*

Average count of offender races in U.S.A.

```python
offender = data.groupby('OFFENDER_RACE', as_index = False).median()
plt.figure(figsize=(20, 10))
sns.barplot(data= data, x= 'OFFENDER_RACE', y= 'TOTAL_OFFENDER_COUNT')
plt.show()
```

*from the above representation of median value of total offender count for each offender race we can observe that multiple races were involved in hate crime followed Native Hawaiian or Other Pacific Islanders,Black or African American,American Indians or Alaska Native*

```python
plt.figure(figsize= (50, 8))
data['AGENCY_TYPE_NAME'].value_counts().plot.pie(autopct= '%1.2F%%', fontsize=8, startangle= 240,  shadow= True)
plt.show()
```

*from the above pie chart representation we can conclude that most of the hate crime offend complaints were filled in city agencies*

```python
offender = data.loc[data['OFFENDER_RACE'].isin(['Unknown', 'American Indian or Alaska Native', 'White','Black or African American','Asian','Native Hawaiian or Other Pacific Islander'])]
offender_type = offender.groupby('OFFENDER_RACE', as_index= False).mean()

fig = plt.figure(figsize= (40,30))
axs = fig.subplots(nrows=1, ncols= 2)
sns.barplot(data = offender_type, x= 'OFFENDER_RACE', y = 'ADULT_OFFENDER_COUNT', ax = axs[0], palette= 'magma')
sns.barplot(data = offender_type, x= 'OFFENDER_RACE', y = 'JUVENILE_OFFENDER_COUNT', ax = axs[1], palette = 'viridis')
plt.show()
```

*from the above representation we can observe that Adults from American Indian race or Alaska Natives were offenders whereas Juvenile of Asian race were offenders for hate crime*
